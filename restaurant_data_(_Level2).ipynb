# ***Geographic Analysis***

```python
df = pd.read_csv('/content/drive/MyDrive/Copy of Dataset .csv')
df
```

```python
import pandas as pd
import os
import folium
from IPython.display import display
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
restaurant_names = df["Restaurant Name"]
latitude = df["Latitude"]
longitude = df["Longitude"]
```

```python
x = df[["Latitude","Longitude"]]
num_clusters = 5
```

```python
map_centre = [latitude.mean(),longitude.mean()]
restaurants_map = folium.Map(location = map_centre,zoom_starts = 12)
```

```python
cluster_color = ["Purple","rerd","blue","Green","black"]
```

```python
for index , row in df.iterrows():
  restaurants_name = row["Restaurant Name"]
  latitude = row["Latitude"]
  longitude = row["Longitude"]
  cuisines = row["Cuisines"]
  rating = row["Aggregate rating"]

```

```python
popup_text = f"Restaurant:(restaurant_name)\nCuisines:{cuisines}|{rating}"
```

```python
marker = folium.Marker([latitude,longitude],popup = popup_text)
```

```python
marker.add_to(restaurants_map)
```
