# ***Price Range Distribution***

```
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Copy of Dataset .csv')
df
```

```
prize_range_count = df['Price range'].value_counts().sort_index()
prize_range_count
```

```
total_res = len(df)
percentage_of_price_range = (prize_range_count/total_res)*100
percentage_of_price_range
```

```
plt.figure(figsize = (8,6))
plt.bar(prize_range_count.index,prize_range_count.values,color = ['turquoise'])
plt.xlabel("Price Range")
plt.ylabel("no. of restaurants")
plt.title('Distribution of price ranges the res')
plt.show()
```

```
percentage_of_price_range = (prize_range_count/total_res)*100
print("percentage in each price range")
percentage_of_price_range
```

# ***Online Delivery***

```
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Copy of Dataset .csv')
df
```

```
total_res = len(df)
online_res_delivery= len(df[df['Has Online delivery']==1])
percentage_online_delivery = (online_res_delivery/total_res)*100
print("percentage restaurants offering online delivery")
print(percentage_online_delivery)
```

```
avg_rating_withdelivery = df[df['Has Online delivery']==1]['Aggregate rating'].mean()
avg_rating_withoutdelivery = df[df['Has Online delivery']==0]['Aggregate rating'].mean()
print("print avg rating of restaurants with online delivery")
print(avg_rating_withdelivery)
print("print avg rating of restaurants without online delivery")
print(avg_rating_withoutdelivery)
```
